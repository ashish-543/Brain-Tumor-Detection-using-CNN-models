```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Models': ['VGG-16', 'VGG-19', 'ResNet-50', 'Inception-v3', 'Xception'],
    'Precision': [0.98, 0.98, 0.98, 0.98, 0.98],
    'Recall': [0.96, 0.96, 0.97, 0.95, 0.94],
    'F1-Score': [0.97, 0.97, 0.97, 0.97, 0.96]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to long format
df_melted = df.melt(id_vars='Models', var_name='Metrics', value_name='Score')

# Create a grouped bar plot with Metrics on x-axis and Models as hue
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Metrics', y='Score', hue='Models', data=df_melted, palette='Set2')

# Add title and labels
plt.title('Model Performance (Glioma): Precision, Recall, and F1-Score', fontsize=16)
plt.xlabel('Metrics', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.ylim(0.85, 1.0)

# Annotate the values on the bars
for p in ax.patches:
    height = p.get_height()
    if not pd.isna(height):
        ax.annotate(f'{height:.2f}',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom',
                    xytext=(0, 4),
                    textcoords='offset points',
                    fontsize=9)

# Add legend
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
